**Paso 1** Importar las librerias necesarias para la actividad.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import datetime
import time
import matplotlib.pyplot as plt
```

**Paso 2** Cargamos el fichero sobre el que queremos operar.

```python
df = pd.read_csv("datos/pedidos_de_venta.csv")
```

**Paso 3** Mostramos nuestro dataframe que contiene los datos del fichero csv.

```python
df
```

**Paso 4** Comprobamos los tipos de las columnas que hemos importado.

```python
df.dtypes
```

**Paso 5** Convertimos la columna object 'Fecha de pedido' a datetime.

```python
df['Fecha de pedido'] = pd.to_datetime(df['Fecha de pedido'])
```

**Paso 6** Comprobamos el resultado.

```python
df.dtypes
```

```python
df
```

**Paso 7** Creamos 2 columnas donde se recojan el Año y el Mes de la fecha de pedido.

```python
df['Year'] = df['Fecha de pedido'].dt.year
df['Month'] = df['Fecha de pedido'].dt.month
```

```python
df
```

**Paso 8** Agrupamos por meses y sumamos los totales

```python
data = df.groupby(by=["Month"], as_index = False)["Total"].sum()
```

**Paso 9** Mostramos el resultado

```python
data
```

**Paso 10** Creamos un gráfico lineal

```python
sns.set(rc={'figure.figsize':(12,12)})
sns.lineplot(data["Month"], data["Total"], size_order= 1)
```

**Paso 11** Construimos un diagrama de barras

```python
plt = sns.barplot(x = data['Month'], y = data['Total'])
plt.set(xlabel = "Mes", ylabel = "Pedidos €")
```
